### Step 12. How talking_about_count correlated with checkins for BuschGardens?

```python
data_BuschGardens=data.loc[data.dataset_id==55545]
data_BuschGardens = data_BuschGardens.set_index(pd.to_datetime(data_BuschGardens['time']))

checkin_aggMonth=data_BuschGardens.resample('M').max()[['talking_about_count','checkins']]
checkin_month = checkin_aggMonth.diff()

print (checkin_aggMonth.corr())
print (checkin_month.corr())
import matplotlib.pyplot as plt
%matplotlib inline  

checkin_month.plot()
plt.savefig('seris.png')
```

We can see there is a similar trend between # talking_about_count and # checkins in BuschGardens. But we need to figure out how long is the delay. 

```python
print (pd.rolling_corr(checkin_month.iloc[:,0],checkin_month.iloc[:,1],2,pairwise=True))
```

The rolling correlations for window size = two months are very significant.

```python
checkin_month['pre_talk']=checkin_aggMonth.talking_about_count.shift(1)
checkin_month['pre_talk_2']=checkin_aggMonth.talking_about_count.shift(2)
checkin_month['pre_talk_3']=checkin_aggMonth.talking_about_count.shift(3)
checkin_month['sum_pre']=checkin_month['pre_talk_2']+checkin_month['pre_talk_3']
```

### Step 13. Plot the regression model with the sum of # talking about count in 2-3 month ago.

```python

import seaborn as sns
#sns.regplot(x="talking_about_count", y="checkins", data=checkin_aggMonth);
sns.jointplot(x="sum_pre", y="checkins", data=checkin_month, kind="reg");
```

```python
from sklearn.linear_model import LinearRegression
linear=checkin_month.reset_index()
linear  = linear[4:]
import statsmodels.api as sm
model = sm.OLS(endog=linear['checkins'], exog=linear[['pre_talk','pre_talk_2']])
results = model.fit()
print(results.summary())
```
